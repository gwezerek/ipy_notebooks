```python
## Open our final JSON
beerFrame = pd.read_csv('/Users/Gus/Dropbox/FastCo/projects_static/untappd/pumpkinBeers.csv', encoding='utf-8')
```

```python
## Set some vars
latest = 0
beerJSON = []
url = 'http://api.untappd.com/v4/beer/info/'
ourParams = {
    'client_id':'C0B05557830DDCDF519E0979BBEC4DC7F9E6CB1D', 
    'client_secret':'29B78C7E0846640D4359CC691D2CD1145C5EF5AA'
}
```

```python
## Pickling function to make sure we don't lose our work
def pickleJSON(toPickle):
    jsonArrFileOut = open("/Users/Gus/Dropbox/FastCo/projects_static/untappd/individual_beer_json.txt", "wb")
    pickle.dump(toPickle, jsonArrFileOut)
    jsonArrFileOut.close()
```

```python
## Initing a new var to catch the IDs that err out
errIds = []
```

```python
## Get the first 100 beer.bid from the beerFrame
## Run the correct beer info API search for that beer
## Store the bits we want from the resulting JSON in an array
## Repeat an hour after the first batch finished with the next batch

def getVintages(r):
    vintArr = []
    if r.get('vintages', {}).get('items'):
        for beer in r['vintages']['items']:
            vintArr.append(beer['beer']['bid'])
        return vintArr
    else: 
        return None

def getBeerRating():
    global latest, beerJSON, errIds
    for beer in  range(0, len(beerFrame)):
        beerID = beerFrame['beer.bid'][latest]
        ourRequest = requests.get(url + str(beerID), params = ourParams)
        if ourRequest.status_code == 200:
            try: 
                r = ourRequest.json()['response']['beer']
            except ValueError:
                errIds.append(latest)
            else:
                singleBeer = {
                    'bid' : r['bid'],
                    'beer_name' : r['beer_name'],
                    'is_in_production' : r['is_in_production'],
                    'is_homebrew' : r['is_homebrew'],
                    'rating_count' : r['rating_count'],
                    'rating_score' : r['rating_score'],
                    'vintages' : getVintages(r)
                }
                beerJSON.append(singleBeer)
            latest += 1
        else:
            print(latest, time.strftime("%Y-%m-%d %H:%M"))
            pickleJSON(beerJSON)
            time.sleep(3600)
    return beerJSON
        
finalBeerJSON = getBeerRating()
print(len(finalBeerJSON))  ## Should be 2278 - len(errIds)
pickleJSON(finalBeerJSON)
```

```python
## New pickling function to make sure we don't lose our work
def pickleJSONExtras(toPickle):
    jsonArrFileOut = open("/Users/Gus/Dropbox/FastCo/projects_static/untappd/individual_beer_json_extras.txt", "wb")
    pickle.dump(toPickle, jsonArrFileOut)
    jsonArrFileOut.close()
```

```python
## Loop through the errIds
## Try to append them
## If it works remove them from the errIds array
## Keep running until the length of the errId array is 0

def getRemainingBeer():
    global beerJSON, errIds
    while len(errIds) > 0:
        for beerID in errIds:
            ourRequest = requests.get(url + str(beerID), params = ourParams)
            if ourRequest.status_code == 200:
                try: 
                    r = ourRequest.json()['response']['beer']
                except ValueError:
                    continue
                else:
                    singleBeer = {
                        'bid' : r['bid'],
                        'beer_name' : r['beer_name'],
                        'is_in_production' : r['is_in_production'],
                        'is_homebrew' : r['is_homebrew'],
                        'rating_count' : r['rating_count'],
                        'rating_score' : r['rating_score'],
                        'vintages' : getVintages(r)
                    }
                    beerJSON.append(singleBeer)
                    errIds.remove(beerID)
                    time.sleep(5)
            else:
                print(len(errIds), time.strftime("%Y-%m-%d %H:%M"), errIds)
                pickleJSONExtras(beerJSON)
                time.sleep(3600)
    return beerJSON
        
finalBeerJSON = getRemainingBeer()
print(len(finalBeerJSON))  ## Should be 2278 - len(errIds)
pickleJSONExtras(finalBeerJSON)
```

```python
## Alright, so we still have 20 ids that are erroring out
## But we're just going to drop those from the data--need to move on

finalBeerJSON = beerJSON
print(len(finalBeerJSON))  ## Should be 2278 - len(errIds)
pickleJSONExtras(finalBeerJSON)
```
