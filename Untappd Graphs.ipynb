```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
## Open our final JSON
df = pd.read_csv('/Users/Gus/Dropbox/FastCo/projects_static/untappd/pumpkinBeers_final.csv', encoding='utf-8', index_col=0)
df.head(1)
```

```python
df['rating_score'].hist(bins=50)
```

```python
## Looks like we have a number of beers with no ratings
## Let's mask those

has_ratings = df[df['rating_count'] != 0]
has_ratings['rating_score'].hist(bins=50)
```

```python
has_ratings = has_ratings.drop('is_homebrew',1)
has_ratings.describe()
```

```python
## Seems like every one of these beers is in production, or more likely, that the column is bogus
## Let's toss it

has_ratings = has_ratings.drop('is_in_production',1)
has_ratings.describe()
```

```python
## Same with 'brewery_active'
## Let's toss it

has_ratings = has_ratings.drop('brewery_active',1)
has_ratings.describe()
```

```python
## Also let's toss the ones where abv == 0 (probly wasn't provided)
has_ratings = has_ratings[has_ratings.beer_abv > 0]
has_ratings.describe()
```

```python
plt.scatter(has_ratings['beer_abv'], has_ratings['rating_score'])
plt.xlabel("ABV")
plt.ylabel("Rating")
```
