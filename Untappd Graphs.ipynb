```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt

# Set some Pandas options
pd.set_option('display.max_columns', 25)
pd.set_option('display.max_rows', 25)
```

```python
## Open our final JSON
df = pd.read_csv('/Users/Gus/Dropbox/FastCo/projects_static/untappd/pumpkinBeers_final.csv', encoding='utf-8', index_col=0)
df.head(1)
```

```python
## Looks like we have a number of beers with no ratings
## Let's mask those

has_ratings = df[df['rating_count'] != 0]
```

```python
has_ratings = has_ratings.drop('is_homebrew',1)
```

```python
## Seems like every one of these beers is in production, or more likely, that the column is bogus
## Let's toss it

has_ratings = has_ratings.drop('is_in_production',1)
has_ratings.describe()
```

```python
## Same with 'brewery_active'
## Let's toss it

has_ratings = has_ratings.drop('brewery_active',1)
has_ratings.describe()
```

```python
## Also let's toss the ones where abv == 0 (probly wasn't provided)
has_ratings = has_ratings[has_ratings.beer_abv > 0]
has_ratings.describe()
```

```python
plt.scatter(has_ratings['beer_abv'], has_ratings['rating_score'])
plt.xlabel("ABV")
plt.ylabel("Rating")
```

```python
## Finding the optimal number of bins for the histogram
sturges = lambda n: int(log2(n) + 1)
square_root = lambda n: int(sqrt(n))
doanes = lambda data: int(1 + log(len(data)) + log(1 + scipy.stats.kurtosis(data) * (len(data) / 6.) ** 0.5))

n = len(has_ratings)
sturges(n), square_root(n), doanes(has_ratings['rating_count'])
```

```python
has_ratings['rating_score'].hist(bins=doanes(has_ratings['rating_count']))
```

```python
## Not a "True" Bayesian estimator, as the threshold of 30 for m is arbitrary
## But given the distribution of scores, this should help tamper down the ratings with fewer votes
## See the bottom of http://www.imdb.com/chart/top for the same methodology

R = has_ratings['rating_score']
v = has_ratings['rating_count']
m = 30
C = 3.595751
has_ratings['weighted_score'] = (v / (v+m)) * R + (m / (v+m)) * C
has_ratings.head(10)
```

```python
## Open the version with weighted scores
has_ratings = pd.read_csv('/Users/Gus/Dropbox/FastCo/projects_static/untappd/pumpkinBeers_has_ratings.csv', encoding='utf-8', index_col=0)
has_ratings
```

```python
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```python
pd.set_option('display.max_rows', 50)

top_by_state = has_ratings.sort('weighted_score', ascending=False).groupby('brewery_state', as_index=False).head(3)
top_us_states = top_by_state[top_by_state['brewery_state'].isin(states)]
top_us_states.sort('weighted_score', ascending=False, inplace=True)
top_us_states.head(3)
```
